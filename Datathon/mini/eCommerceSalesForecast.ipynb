# Sales by Brand by Month

- df: monthly_sales_by_brand

```python
# Groupby Brands the sales rec in train


# Extract the first six columns which are not dates
non_date_columns = train.columns[:6]

# Convert the remaining columns (which are dates) to datetime format
date_columns = pd.to_datetime(train.columns[6:])

# Correctly reassign the columns back to the DataFrame
train.columns = non_date_columns.tolist() + date_columns.strftime('%Y-%m-%d').tolist()

# Group by '브랜드' column and sum the sales across all items for each brand
sales_by_brand = train.groupby('브랜드').sum().reset_index()

sales_by_brand.head()
```

```python
# melt the df to Long Format
# Assuming 'sales_by_brand' is your DataFrame and it has been corrected to have only sales data columns in datetime format
# Melt the DataFrame
for_melt = sales_by_brand.drop(['ID', '제품', '대분류', '중분류', '소분류'], axis=1)
sales_long = pd.melt(for_melt, id_vars=['브랜드'], var_name='Date', value_name='Sales')

# Convert 'Date' from string to datetime
sales_long['Date'] = pd.to_datetime(sales_long['Date'])

```

```python
# Group by Brand and Month

# Extract year and month from 'Date' for grouping
sales_long['YearMonth'] = sales_long['Date'].dt.to_period('M')

# Group by '브랜드' and 'YearMonth', then sum sales
monthly_sales_by_brand = sales_long.groupby(['브랜드', 'YearMonth']).agg({'Sales':'sum'}).reset_index()

# If you need to convert 'YearMonth' back to a datetime format for plotting or other purposes:
monthly_sales_by_brand['YearMonth'] = monthly_sales_by_brand['YearMonth'].dt.to_timestamp()

monthly_sales_by_brand.head()
```

```python
pd.set_option('display.max_colwidth', 100)

train.merge(words, on='제품', how='inner')

tag = train.merge(words, on='제품', how='inner')
# Assuming 'train' is your DataFrame with '소분류' and '제품' columns
# And 'words' is your DataFrame with '제품' and 'Keywords' columns
# And you've already merged them as in your example 'tag'

# Group by '소분류' and join the keywords into a single string
keywords_by_category = tag.groupby('소분류')['Keywords'].agg(lambda x: ' '.join(x)).reset_index()

keywords_by_category
```

```python
import pandas as pd

# Assuming 'keywords_by_category' is your DataFrame with '소분류' and 'Keywords' columns,
# where 'Keywords' contains all keywords for that category separated by spaces.

# Define a function to count keywords within each category and find the top one
def get_top_keyword_per_category(df):
    # Dictionary to store the top keyword for each category
    top_keywords = {}
    
    # Iterate over each category
    for category in df['소분류'].unique():
        # Get all keywords for the category
        category_keywords = df[df['소분류'] == category]['Keywords'].str.cat(sep=' ').split()
        
        # Count the occurrences of each keyword
        keyword_counts = pd.Series(category_keywords).value_counts()
        
        # Get the top keyword
        top_keyword = keyword_counts.idxmax()
        
        # Store the top keyword for this category
        top_keywords[category] = top_keyword
    
    # Convert the dictionary to a DataFrame
    top_keywords_df = pd.DataFrame(list(top_keywords.items()), columns=['소분류', 'TopKeyword'])
    
    return top_keywords_df

# Apply the function to your DataFrame
top_keywords_df = get_top_keyword_per_category(keywords_by_category)

# Display the result
print(top_keywords_df)
```
